# Krypto Puzzle Solver

This notebook implements a solver for Krypto-style arithmetic puzzles using recursive search.

## Sample Puzzle

```python
numbers = [4, 11, 17, 23, 29]
target = 18
```

## Solver Implementation

```python
def solve_krypto(nums, target):
    solutions = set()
    
    def helper(numbers, expressions):
        if len(numbers) == 1:
            if abs(numbers[0] - target) < 1e-6:
                solutions.add(expressions[0])
            return
        
        for i in range(len(numbers)):
            for j in range(len(numbers)):
                if i != j:
                    a, b = numbers[i], numbers[j]
                    exp_a, exp_b = expressions[i], expressions[j]
                    
                    remaining_nums = [numbers[k] for k in range(len(numbers)) if k not in (i, j)]
                    remaining_exp = [expressions[k] for k in range(len(numbers)) if k not in (i, j)]
                    
                    results = [
                        (a + b, f"({exp_a} + {exp_b})"),
                        (a - b, f"({exp_a} - {exp_b})"),
                        (a * b, f"({exp_a} * {exp_b})")
                    ]
                    
                    if b != 0:
                        results.append((a / b, f"({exp_a} / {exp_b})"))
                    
                    for val, expr in results:
                        helper(remaining_nums + [val], remaining_exp + [expr])
    
    helper(nums, [str(n) for n in nums])
    return list(solutions)
```

## Running the Solver

```python
solutions = solve_krypto(numbers, target)

print(f"Unique solutions found: {len(solutions)}\n")

for s in solutions[:10]:
    print(s)
```

## Determining Difficulty

```python
def classify_difficulty(num_solutions):
    if num_solutions > 20:
        return "Easy"
    elif num_solutions > 5:
        return "Medium"
    elif num_solutions > 1:
        return "Hard"
    elif num_solutions == 1:
        return "Expert"
    else:
        return "No Solution"
```

```python
difficulty = classify_difficulty(len(solutions))

print("Difficulty:", difficulty)
```

## Testing Multiple Puzzles

To evaluate the solver, several Krypto-style puzzles were tested.  
For each puzzle, the solver returns the number of valid solutions and assigns a difficulty level based on the size of the solution set.

```python
test_cases = [
    ([4, 11, 17, 23, 29], 18),
    ([7, 11, 18, 23, 29], 15),
    ([2, 5, 10, 20, 25], 17),
    ([3, 6, 9, 12, 24], 21)
]

for nums, tgt in test_cases:
    solutions = solve_krypto(nums, tgt)
    difficulty = classify_difficulty(len(solutions))
    
    print(f"Numbers: {nums}, Target: {tgt}")
    print(f"Unique solutions found: {len(solutions)}")
    print(f"Difficulty: {difficulty}")
    
    if solutions:
        print("Example solution:", solutions[0])
    else:
        print("No solution found.")
    
    print("-" * 50)
```

## Random Puzzle Generation

```python
import random

def generate_puzzle():
    numbers = random.sample(range(1, 31), 5)
    target = random.randint(1, 30)
    return numbers, target

nums, tgt = generate_puzzle()
print("Generated puzzle:")
print("Numbers:", nums)
print("Target:", tgt)

solutions = solve_krypto(nums, tgt)
print(f"Unique solutions found: {len(solutions)}")
print("Difficulty:", classify_difficulty(len(solutions)))
```
